```python
import json
import torch
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForCausalLM

def read_jsonl(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        return [json.loads(line) for line in f]
    
    
# data = read_jsonl("train_data.jsonl")
data = read_jsonl("converted_dataset.jsonl")
dataset = Dataset.from_list(data)

token = ""  
model_name = "ruslandev/llama-3-8b-gpt-4o-ru1.0"

tokenizer = AutoTokenizer.from_pretrained(model_name, use_auth_token=token)
model = AutoModelForCausalLM.from_pretrained(model_name, use_auth_token=token, torch_dtype=torch.float16).to("cuda")


if tokenizer.pad_token_id is None:
    tokenizer.pad_token_id = tokenizer.eos_token_id

def tokenize_function(example):
    inputs = tokenizer(
        example["prompt"], 
        max_length=77, 
        truncation=True, 
        padding="max_length"
    )
    labels = tokenizer(
        example["response"], 
        max_length=77, 
        truncation=True, 
        padding="max_length"
    )

    inputs["labels"] = labels["input_ids"]
    return inputs


dataset = dataset.map(tokenize_function, batched=True)
dataset_split = dataset.train_test_split(test_size=0.3, seed=42)

train_dataset = dataset_split["train"]
eval_dataset = dataset_split["test"]

```

```python
from peft import get_peft_model, LoraConfig, TaskType
from transformers import Trainer, TrainingArguments


lora_parameters = {
    'lora_r': 8,
    'target_modules': ["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],
    'lora_alpha': 16,
    'lora_dropout': 0,
    'lora_bias': "none",
    'lora_use_gradient_checkpointing': "unsloth",
    'lora_random_state': 42,
}

peft_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    inference_mode=False,
    r=lora_parameters['lora_r'],
    lora_alpha=lora_parameters['lora_alpha'],
    lora_dropout=lora_parameters['lora_dropout'],
    target_modules=lora_parameters['target_modules']
)

peft_model = get_peft_model(model, peft_config)

```

```python
peft_model.print_trainable_parameters()
```

```python
training_args = TrainingArguments(
    output_dir="outputs", 
    per_device_train_batch_size=3,  
    num_train_epochs=2, 
    logging_dir="./logs",  
    logging_steps=2, 
    save_strategy="epoch",  
    # gradient_accumulation_steps=4, 
    learning_rate=1e-4,  
    weight_decay=0.01, 
    seed=3407,  
    eval_strategy="steps"
)


trainer = Trainer(
    model=peft_model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset
)

trainer.train()
```

```python
system_prompt = "system: Ты чат-бот, который генерирует диалоги. Продолжи смешной анекдот.\nuser:"
user_prompt = "Заходит как-то улитка в бар:"
full_prompt = f"{system_prompt}{user_prompt}"

inputs = tokenizer(full_prompt, return_tensors="pt", padding=True, truncation=True, max_length=77).to('cuda')

generated = peft_model.generate(
    input_ids=inputs['input_ids'],
    attention_mask=inputs['attention_mask'],
    max_length=150,
    num_return_sequences=1,
    temperature=0.6,
    top_p=0.9,
    do_sample=True, 
    pad_token_id=tokenizer.eos_token_id,  
    repetition_penalty=1.05 
)

print(tokenizer.decode(generated[0], skip_special_tokens=True))
```

```python
# Сохранение модели и конфигурации LoRA
output_dir = "./model/"  # Укажите директорию, куда сохранить модель
peft_model.save_pretrained(output_dir)  # Сохраняем модель

# Сохранение конфигурации LoRA
peft_config.save_pretrained(output_dir)

# Сохранение токенизатора (если необходимо)
tokenizer.save_pretrained(output_dir)
```
